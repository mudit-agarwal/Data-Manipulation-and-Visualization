#### 1) Simple Plotting

```python
import matplotlib.pyplot as plt

x=[1,3,5,7]
y=[2,4,6,8]

#This creates a dot at the (x,y) co-ordinate
plt.scatter(x,y)

#This creates the line between two points
plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,3,5,7]
y=[2,4,6,8]

x1=[2,5,8,4]
x2=[8,5,7,1]

plt.scatter(x,y)
plt.plot(x,y)

#We can plot multiple lines as well. Here"r--" means red colour with dashes.
#Similarly other stuff can also be done
plt.scatter(x1,x2)
plt.plot(x1,x2,"r--")

plt.show()
```

#### 2) Plot a Function y=x^3

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,5,0.1)
y = x**3
plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

Note : If you only give X and not Y, then Y is taken as 1,2,3....

### 4) Labelling of Graph 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,5,0.1)
y = x**3
y2 = x**2

# How to make changes in the lines of the graph
plt.plot(x,y,color="Blue",linewidth=5,label="x^3 Graph")
plt.plot(x,y2,color="Red",linewidth=5,label="x^2 Graph")

# Now to label the axis,graphs and legends
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.title("A Sample Matplotlib Graph")

# Too add text in the graph somewhere
plt.text(3,72,"I'm highlighting this point",fontsize=8)

plt.grid()
plt.show()
```

#### 5) Bubble Graph

```python
import numpy as np
import matplotlib.pyplot as plt

year=[2015,2016,2017,2018,2019]
salary=[100,120,140,160,180]
population=[20,40,60,80,100]

c = np.arange(0,len(population))
plt.scatter(year,salary,s=population,c=c)
plt.title("Bubble Graph")

plt.show()
```

#### 6) Pie-Graphs

```python
labels=["A","B","C","D"]
sizes=[3,4,6,2]
colors=["Red","Blue","Green","Yellow"]
explode=[0,0,0,0.2]
plt.title("Pictographical Pie Chart")
plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct="% .2f%%",startangle=90,counterclock=False)
#(sizes of pie,colors,explode is to make one part of the pie come out,labels is labbelling, autopct is for showing the %)
#(start angel is for where graph should start, counterclock is clockwise)

plt.axis("equal")
plt.show()
```

#### 7) Bar Graphs

```python
import numpy as np
import matplotlib.pyplot as plt

year=[2015,2016,2017,2018,2019]
salary=[100,120,140,160,180]
population=[20,40,60,80,100]

plt.bar(year,population,width=0.4)

plt.show()
```
